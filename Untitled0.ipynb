```python
import talib as ta
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
import datetime as dt
import yfinance as yf
```

```python
start = dt.datetime(2022, 1, 1)
end = dt.datetime.now()
data = yf.download("AAPL", start=start, end=end)
```

```python
data
```

```python
data['SMA_100'] = ta.SMA(data['Close']['AAPL'].values, 100)
```

```python
data
```

```python
plt.plot(data["Close"])
plt.plot(data["SMA_100"])
```

```python
data['EMA_100'] = ta.EMA(data['Close']['AAPL'].values, 100)
```

```python
plt.plot(data["Close"])
plt.plot(data["EMA_100"])
```

```python
data['RSI'] = ta.RSI(data['Close']['AAPL'].values)
```

```python
fig, axs = plt.subplots(2, 1, gridspec_kw={"height_ratios": [3, 1]}, figsize=(10,6))

axs[0].plot(data['Close'])
axs[1].axhline(y=70, color="r", linestyle="--")
axs[1].axhline(y=30, color="g", linestyle="--")
axs[1].plot(data['RSI'], color="orange")
```

```python
data['LINEARREG'] = ta.LINEARREG(data['Close']['AAPL'].values, 200)
```

```python
plt.plot(data["Close"])
plt.plot(data["LINEARREG"])
```

```python
start = dt.datetime(2022, 1, 1)
end = dt.datetime.now()

data_gs = yf.download("GS", start=start, end=end)
macd, macd_signal, macd_hist = ta.MACD(data_gs['Close']['GS'].values)

fig, axs = plt.subplots(2, 1, gridspec_kw={"height_ratios": [3, 1]}, figsize=(10,6))
c = ["red" if cl < 0 else "green" for cl in macd_hist]

axs[0].plot(data_gs['Close']['GS'])
axs[1].plot(macd, 'b-')
axs[1].plot(macd_signal, '--', color="orange")
axs[1].bar(range(len(macd_hist)), macd_hist, color=c)
axs[1].set_xticks(range(0, len(macd_hist), 50))
axs[1].set_xticklabels(data_gs.index[range(0, len(macd_hist), 50)].strftime('%Y-%m-%d'), rotation=45)
```

```python
start = dt.datetime(2022, 1, 1)
end = dt.datetime.now()

data = yf.download("TSLA", start=start, end=end)
```

```python
data['ENGULFING'] = ta.CDLENGULFING(data['Open']['TSLA'].values, data['High']['TSLA'].values, data['Low']['TSLA'].values, data['Close']['TSLA'].values)
data
```

```python
import mplfinance as mpf
import pandas as pd

plot_data = data.copy()
plot_data.columns = plot_data.columns.droplevel(1)

fig, axs = plt.subplots(2, 1, gridspec_kw={"height_ratios": [3, 1]}, figsize=(16,12))
colors = mpf.make_marketcolors(up='g', down='r')

mpf_style = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=colors)
mpf.plot(plot_data, type="candle", ax=axs[0], style=mpf_style)
axs[1].plot(plot_data['ENGULFING'], color="blue")
```
